```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
plt.title('Iris Dataset')
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
```

```python
X = df[['petal_length', 'petal_width']]
y = df['species']
```

```python
k = 8
knn_classifier = KNeighborsClassifier(n_neighbors=k)
knn_classifier.fit(X, y)

new_data = pd.DataFrame({'petal_length': [5.5], 'petal_width': [2.1]}) 
```

```python
predicted = knn_classifier.predict(new_data)

print(f"Predicted Species: {predicted[0]}")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))

sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', ax=ax)
sns.scatterplot(x=new_data['petal_length'], y=new_data['petal_width'], color='red', marker='o', s=100, label='predicted', ax=ax)
ax.legend()

plt.grid()
```
